```python
#Creo mi tienda como una clase. 
class TiendaOnline:
    def __init__(self, nombre): 
        self.nombre = nombre
        self.inventario = [] # como es una lista, uso corchetes, y adentro va a tener diccionarios de los productos
        self.ventas_totales= 0.0 # le pongo el cero con el punto y el cero, para que sea float. o sea que incluya numeros positivos, negativos y decimales
        self.valor_inventario=0

    def agregar_producto (self, nombre, precio, cantidad): #le pongo self, porque es una "funcion" dentro de una clase.
        for producto in self.inventario: #itero sobre los diccionarios de productos que estan dentro de la lista de inventario 
            if producto["nombre"] == nombre:  # compruebo si el nombre ya existe en el diccionario productode cada producto
                producto["cantidad"] += cantidad  # Si el producto ya existe, actualizo la cantidad.
                print(f"Cantidad actualizada para '{nombre}'. Nueva cantidad: {producto['cantidad']}")
                return
        #Si no existe, agrego un nuevo producto al inventario
        nuevo_producto = {"nombre": nombre, "precio": precio, "cantidad": cantidad}
        self.inventario.append(nuevo_producto)
        print(f"Producto '{nombre}' agregado al inventario. Precio: ${precio}, Cantidad: {cantidad}")

    def ver_inventario(self):
        for producto in self.inventario:  #uso bucle for para recorrer el inventario.
            nombre = producto["nombre"]
            precio = producto["precio"]
            cantidad = producto["cantidad"]
            print(f"Inventario de productos: Nombre: {nombre}, Precio: ${precio:}, Cantidad: {cantidad}") #Uso {precio:.2f} para que se muestre con 2 decimales (2 posiciones y f flotante)

    def buscar_producto (self, nombre):
        for producto in self.inventario: 
            if nombre in producto["nombre"]: #chequeo que el nombre este dentro de la variable producto que es un diccionario producto
                print(f"Nombre: {producto['nombre']} - Precio: ${producto['precio']} - Cantidad: {producto['cantidad']}") #Si encuentra el producto, imprime sus detalles.

    def actualizar_stock(self, nombre, cantidad):
        for producto in self.inventario:
            if producto["nombre"] == nombre: # Comparo el nombre del producto si es que ya existe en la posicion 0 de productos, que son los nombres, dentro de los diccionarios.
                producto["cantidad"] += cantidad
                if producto["cantidad"] < 0:
                    producto["cantidad"] = 0 # Evito que la cantidad sea negativa
                print(f"Stock de '{nombre}' actualizado. Nueva cantidad: {producto['cantidad']}")
                return
        print(f"No se encontro '{nombre}' en el inventario.") #Si el producto no esta en el inventario muestro un mensaje.

    def eliminar_producto(self, nombre):
        for producto in self.inventario:
            if producto["nombre"] == nombre: #Busco el producto por nombre.
                self.inventario.remove(producto) #Elimino el producto del inventario si es que ya existe.
                print(f"El producto '{nombre}' ha sido eliminado del inventario.")
                return
        print(f"No se encontró '{nombre}' en el inventario.") #si no esta en el inventario, imprimo un mensaje

    def calcular_valor_inventario(self):
        for producto in self.inventario:
            self.valor_inventario += producto["precio"] * producto["cantidad"] #Multiplico el precio de cada producto por la cantidad que tengo.
        print(f"EL valor total de toda la mercancia de Verduleria y Fruteria Eleonora es: ${self.valor_inventario}")


```

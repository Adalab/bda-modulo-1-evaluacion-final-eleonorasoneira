```python
#Creo mi tienda como una clase. 
class TiendaOnline:
    def __init__(self, name): 
        self.name = name
        self.inventory = [] # como es una lista, uso corchetes, y adentro va a tener diccionarios de los productos
        self.total_sales= 0.0 # le pongo el cero con el punto y el cero, para que sea float. o sea que incluya numeros positivos, negativos y decimales
        self.inventory_value=0.0

    def add_product (self, name, price, quantity): #le pongo self, porque es una "funcion" dentro de una clase.
        for product in self.inventory: #itero sobre los diccionarios de productos que estan dentro de la lista de inventario 
            if product["nombre"] == name:  # compruebo si el nombre ya existe en el diccionario productode cada producto
                product["cantidad"] += quantity  # Si el producto ya existe, actualizo la cantidad.
                print(f"Cantidad actualizada para '{name}'. Nueva cantidad: {product['cantidad']}")
                return
        #Si no existe, agrego un nuevo producto al inventario
        new_product = {"nombre": name, "precio": price, "cantidad": quantity}
        self.inventory.append(new_product)
        print(f"Producto '{name}' agregado al inventario. Precio: ${price}, Cantidad: {quantity}")

    def see_inventory(self):
        for product in self.inventory:  #uso bucle for para recorrer el inventario.
            name = product["nombre"]
            price = product["precio"]
            quantity = product["cantidad"]
            print(f"Inventario de productos: Nombre: {name}, Precio: ${price:.2f}, Cantidad: {quantity}") #Uso {precio:.2f} para que se muestre con 2 decimales (2 posiciones y f flotante)

    def look_for_product (self, name):
        for product in self.inventory: 
            if name in product["nombre"]: #chequeo que el nombre este dentro de la variable producto que es un diccionario producto
                print(f"Nombre: {product['nombre']} - Precio: ${product['precio']} - Cantidad: {product['cantidad']}") #Si encuentra el producto, imprime sus detalles.
                return
            
    def stock_update(self, name, quantity):
        for product in self.inventory:
            if product["nombre"] == name: # Comparo el nombre del producto si es que ya existe en la posicion 0 de productos, que son los nombres, dentro de los diccionarios.
                product["cantidad"] += quantity
                if product["cantidad"] < 0:
                    product["cantidad"] = 0 # Evito que la cantidad sea negativa
                print(f"Stock de '{name}' actualizado. Nueva cantidad: {product['cantidad']}")
                return
        print(f"No se encontro '{name}' en el inventario.") #Si el producto no esta en el inventario muestro un mensaje.

    def delete_product(self, name):
        for product in self.inventory:
            if product["nombre"] == name: #Busco el producto por nombre.
                self.inventory.remove(product) #Elimino el producto del inventario si es que ya existe.
                print(f"El producto '{name}' ha sido eliminado del inventario.")
                return
        print(f"No se encontró '{name}' en el inventario.") #si no esta en el inventario, imprimo un mensaje

    def calculate_inventory_value(self):
        for product in self.inventory: #Por cada producto del inventario
            self.inventory_value += product["precio"] * product["cantidad"] #Multiplico el precio de cada producto por la cantidad que tengo.
        print(f"EL valor total de toda la mercancia de Verduleria y Fruteria Eleonora es: ${self.inventory_value:.2f}")

```

```python
print("--------------------------------------------------------")
# Creo una instancia de la clase TiendaOnline
mi_TiendaOnline = TiendaOnline("Verduleria y Fruteria Eleonora")
# Imprimo el nombre de mi tienda solo para que quede mas bonito
print(mi_TiendaOnline.name)
print("--------------------------------------------------------")

# Agrego productos al inventario: con los valores de cada diccionario: nombre, precio y la cantidad
print("El inventario de Verduleria y Fruteria Eleonora es:")
mi_TiendaOnline.add_product ("Manzana", 1, 50)
mi_TiendaOnline.add_product ("Naranja", 2, 50)   
mi_TiendaOnline.add_product ("Banana", 3, 50)
mi_TiendaOnline.add_product ("Frutilla", 4, 50)
mi_TiendaOnline.add_product ("Limon", 5, 50) 
mi_TiendaOnline.add_product ("Kiwi", 6, 50)
mi_TiendaOnline.add_product ("Durazno", 7, 50)  
mi_TiendaOnline.add_product ("Cereza", 8, 50)   
mi_TiendaOnline.add_product ("Sandia", 9, 50)
mi_TiendaOnline.add_product ("Avocado", 10, 50)
mi_TiendaOnline.add_product ("Anana", 11, 50) 
mi_TiendaOnline.add_product ("Melon", 12, 50)
print("--------------------------------------------------------")

#Pruebo agregar algo que ya existe (manzanas), para comprobar que funcione bien el codigo. 
print("Agrego 50 manzanas:")
mi_TiendaOnline.add_product("Manzana", 3.0, 50) 
print("--------------------------------------------------------")

# Busco un producto por nombre, en este caso busco con el nombre Frutilla
print("Búsqueda de productos por nombre. Ej: Frutilla:")
mi_TiendaOnline.look_for_product("Frutilla")
print("--------------------------------------------------------")

## Actualizo el stock de un producto que ya existe, para comprobar quqe el codigo funcione bien.
print("Actualizo el stock de 'Frutilla' (+100):")
mi_TiendaOnline.stock_update("Frutilla", 100)
print("--------------------------------------------------------")

# Actualizo el stock de un producto no existente.
print("Para comprobar codigo actualizo el stock de un producto no existente, en este caso 'Pera'.")
mi_TiendaOnline.stock_update("Pera", 30)
print("--------------------------------------------------------")

# Elimino un producto que si existe
print("Elimino el producto 'Cereza', que si existe en el inventario:")
mi_TiendaOnline.delete_product("Cereza")
print("--------------------------------------------------------")

# Elimino un producto que no existe en el inventario, para comprobar que el codigo funcione
print("Para comprobar codigo intento eliminar un producto que no existe, en este caso: 'Papa'.")
mi_TiendaOnline.delete_product("Papa")
print("--------------------------------------------------------")

#Muestro el inventario actualizado despues de haber hecho varios cambios
print("Inventario Completo:")
mi_TiendaOnline.see_inventory()
print("--------------------------------------------------------")

#Muestro el valor del inventario
mi_TiendaOnline.calculate_inventory_value()
print("--------------------------------------------------------")
```
